```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v=[0, 1, 1, 0, 1, 1, 0, 0, 1, 1, 0]
np.shape(v)
```

```python
dim=100 #number of sample
```

```python
vx=[]
```

```python
#signal koji se moduli≈°e

for i in range(1,11):
    f=np.ones(dim)
    print(v[i])
    x=f*v[i]
    vx=np.concatenate((vx,x))
    print(vx)
```

```python
plt.subplot(3,1,1)
plt.plot(vx)
plt.show()
```

```python
#nosilac informacije

plt.subplot(3,1,2) 
t=np.linspace(0,2,len(vx))
f1=5
w1=2*np.pi*f1
y1=np.cos(2*w1*t)
plt.plot(t,y1)
plt.show()
```

```python
#modulisani signal


plt.subplot(3,1,3) 
mult=vx*y1

t=np.linspace(0,2,len(vx))

plt.plot(t,mult)
plt.show()




```
